```
#Decorador para calcular_tiempo a la función para la cual queremos calcular el tiempo de ejecución(Isabel Vázquez)

from time import time
#Función para calcular el tiempo de ejecución
def calcular_tiempo(f):
   
    def wrapper(*args, **kwargs):     
        inicio = time()       
        resultado = f(*args, **kwargs)       
        tiempo = time() - inicio
        print("Tiempo de ejecución para algoritmo: "+str(tiempo))
        return resultado
    
    return wrapper
from time import time
```

```
#quick_sort

A = [9187, 244, 4054, 9222, 8373, 4993, 5265, 5470, 4519, 7182, 2035, 3506, 4337, 7580, 2554, 2824, 8357, 4447, 7379]

def quick_sort(A):
  if len(A) == 1:
    return A
  elif len(A) == 2:
    return [min(A), max(A)]
  
  IZQ, DER  = [], []
  
  pivote = (A[0]+A[1]+A[2])/3
  
  for i in A:
    IZQ.append(i) if i <= pivote else DER.append(i)
      
  return quick_sort(IZQ) + quick_sort(DER)


#Ejecutar quick_sort para A

@calcular_tiempo
def QS(A):
  return quick_sort(A)
  

print(QS(A))
  
  
```

```
#Cambio Monedas

@calcular_tiempo
def cambio_monedas(CANTIDAD, SISTEMA = [25,10,5,1]):
  SOLUCION = [0 for i in range(len(SISTEMA))]
  VALOR_ACUMULADO = 0
  
  for i in range(len(SISTEMA)):
    monedas = int(( CANTIDAD - VALOR_ACUMULADO)/SISTEMA[i])
    
    SOLUCION[i] = monedas
    VALOR_ACUMULADO += monedas * SISTEMA[i]
    
    
    if CANTIDAD == VALOR_ACUMULADO:
      return SOLUCION;
    
#Ejecutar cambio_monedas para 1201    
print(cambio_monedas(1201))

    
#Ejecutar cambio_monedas para 3555    
print(cambio_monedas(3555))
```

```


def es_prometedora(SOLUCION, ETAPA):
  for i in range(ETAPA+1):
    if(SOLUCION).count(SOLUCION[i])>1 : return False
    
    
    for j in range(i+1, ETAPA + 1):
      if abs(i-j) == abs(SOLUCION[i]- SOLUCION[j]) : return False
      
  return True

def escribe(SOLUCION):
  n = len(SOLUCION)
  for x in range(n):
    print("")
    for i in range(n):
      if SOLUCION[i] == x+1:
        print(" X ", end= "")
      else:
        print(" - ", end= "")
  

def reinas(N, SOLUCION, ETAPA):
  for i in range(1,N+1):
    SOLUCION[ETAPA] = i
    if es_prometedora(SOLUCION, ETAPA):
      if ETAPA == (N -1):
        print("\n\n La solucion es:")
        print(SOLUCION)
        escribe(SOLUCION)
      else:
        reinas(N, SOLUCION, ETAPA + 1)
    else:
      None
    
  SOLUCION[ETAPA] = 0

  
  
#Ejecutar reinas para 8 elementos   
N = 8
SOLUCION = [0 for i in range(N)]
ETAPA = 0  
reinas(N, SOLUCION, ETAPA)
```
